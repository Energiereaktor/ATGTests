```python
# Hier importieren wir die Bibliothek numpy
import numpy as np

#Hier definieren wir die Funktion calculateC
def calculateC(a, b):
    return np.sqrt(a*a+b*b)

# Hier zählen wir a immer um eins hoch und testen dannach b für a.
for a in range(1000):
    for b in range(a*10):
        c=calculateC(a+1, b+1)
        if c % 1 == 0:
            print("Pytagoräische Zahlentripel: a = " + str(a+1) + ", b = " + str(b+1) + ", c = " + str(c) + "!")
```
