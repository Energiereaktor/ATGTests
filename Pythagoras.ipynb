# Generator für Pytagoräische Zahlentripel

#### Numpy
Als erstes importieren wir Numpy und geben der Bibliothek den Namen "np". Diese Bibliothek (Numpy) kann für verschiedene mathematische Operationen und Funktionen gebraucht werden. In diesem Beispiel benutzen wir diese um die Quadratwurzel von a² + b² zu berechnen. Diese Funktion heisst in der Numpy Bibliothek "np.sqrt(x)". Wobei "np" der Name der Bibliothek und "sqrt" der Funktionsname ist. Hier kann man X mit der gewünschten Zahl oder Variabel ersetzen.

#### Funktion: calculateC(a, b)
Als nächstes definieren wir eine Funktion mit dem Namen "calculateC" und geben ihr die Argumente a und b. Dannach berechnen wir die Wurzel von a² + b² und geben diese mit "return" zurück. Um eine Funktion zu definieren brauchen wir das Schlüsselwort "def", welches für "define" also definieren steht. Wir wenden diese Funktion im nächsten Abschnitt an.

#### Schlaufen
Nun haben wir mehrere verschachtelte "for-loops". Als erstes ist dort die Schlaufe "for a in range(1000)". Dies bedeutet, dass wir "a" immer um eins bis 1000 hochzählen und danach den folgenden Code ausführen. Jetzt haben wir eine weitere Schlaufe, in welcher "b" immer um eins hochgezählt wird. In dieser Schlaufe berechnen wir danach "c" für "a" und "b" mit "c=calculateC(a+1, b+1)". Wir fügen zu "a" und "b" +1 hinzu um sicher zu stellen, dass "a" und "b" nie 0 sind. Danach prüfen wir ob "c" eine gerade Zahl ist, indem wir "c" durch 1 dividieren und überprüfen ob der Rest 0 ist. Wenn dem so ist, drucken wir das Ergebnis in die Konsole.

```python
# Hier importieren wir die Bibliothek numpy
import numpy as np

#Hier definieren wir die Funktion calculateC
def calculateC(a, b):
    return np.sqrt(a*a+b*b)

# Hier zählen wir a immer um eins hoch und testen dannach b für a.
for a in range(100):
    for b in range(a*10):
        c=calculateC(a+1, b+1)
        if c % 1 == 0:
            print("Pytagoräische Zahlentripel: a = " + str(a+1) + ", b = " + str(b+1) + ", c = " + str(c) + "!")
```
